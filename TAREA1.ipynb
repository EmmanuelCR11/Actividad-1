```python
# Lista de meses (como cadenas de texto)
precip_months = [
    "jan", "feb", "mar", "apr", "may", "jun",
    "jul", "aug", "sept", "oct", "nov", "dec"
]

# Lista de precipitaciones (como números reales - float)
precip_inches = [
    0.70, 0.75, 1.85, 2.93, 3.05, 2.02,
    1.93, 1.62, 1.84, 1.31, 1.39, 0.84
]
print (precip_months)
print (precip_inches)
```

```python
# Factor de conversión
inch_to_mm = 25.4

# Copia de la lista original
precip_mm = precip_inches.copy()

# Conversión de pulgadas a milímetros
for i in range(len(precip_mm)):
    precip_mm[i] = precip_mm[i] * inch_to_mm
    precip_mm = [value * inch_to_mm for value in precip_inches]
    
    print(precip_mm)
    
```

```python
# Crear lista de listas
all_precip_data = [precip_months, precip_mm]

# Verificación
print(all_precip_data)
```

¿Qué ventajas tiene agrupar datos relacionados en una sola estructura?
Agrupar datos relacionados en una sola estructura, como una lista de listas, me parece muy útil cuando trabajo con datos de océano o geoquímica porque me ayuda a mantener todo organizado y coherente; por ejemplo, asegurar que cada mes corresponda exactamente a su valor de precipitación o que cada muestra de agua esté alineada con su concentración química.

¿Qué limitaciones tiene una lista de listas frente a una tabla?

una lista de listas también se queda corta cuando los análisis se vuelven más complejos, como suele pasar en geoquímica o en estudios de variabilidad oceánica.y tampoco permite hacer cálculos estadísticos o filtrados de manera directa como lo haría una tabla estructurada No tiene nombres de columnas claros, lo que puede generar confusión cuando trabajamos con muchas variables

```python
# Importar la librería necesaria
import matplotlib.pyplot as plt

# Crear figura y ejes (tamaño ajustado)
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar datos usando indexación correcta
ax.bar(
    all_precip_data[0],   # Meses
    all_precip_data[1],   # Precipitación en mm
    color="aqua"
)

# Configurar títulos y etiquetas
ax.set(
    title="Precipitación Media Mensual (mm)",
    xlabel="Mes",
    ylabel="Precipitación (mm)"
)

# Mostrar gráfica
plt.show()
```

¿Por qué es importante saber cómo están organizados los datos antes de graficar?
Es importante saber cómo están organizados los datos antes de graficar porque la estructura determina qué representa cada eje y cómo se relacionan las variables entre sí. Si no se entiende la organización (por ejemplo, qué lista corresponde al eje X y cuál al eje Y, o si los datos están correctamente alineados), se pueden generar gráficos incorrectos

¿Qué limitaciones tiene trabajar con listas frente a usar un DataFrame?
Trabajar con listas tiene varias limitaciones frente a un DataFrame. Las listas no tienen nombres de columnas, lo que dificulta identificar claramente qué representa cada conjunto de datos. Tampoco permiten realizar filtrados, agrupaciones o cálculos estadísticos de forma directa y eficiente. Además, cuando el volumen de datos aumenta o se manejan múltiples variables, las listas se vuelven difíciles de organizar y más propensas a errores.
